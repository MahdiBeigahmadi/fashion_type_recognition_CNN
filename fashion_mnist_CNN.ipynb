```python
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
import tensorflow as tf
import sklearn.datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import random

tf.random.set_seed(42)
(trainingData, trainingLabels), (testingData, testingLabels) = fashion_mnist.load_data()
```

```python
# scale down and normalize the data 
xTrainScaled = trainingData / 255
xTestScaled = testingData / 255
```

```python
classNames = ["T-shirt/top", "Trousers", "Pullover", "Dress", "Coat", "Sandal", "Shirt", "Sneakers", "Bag", "Ankle Boots"]
# Create a new figure
plt.figure(figsize=(15,15))

plt.figure(figsize=(7,7))
for i in range(4):
    ax = plt.subplot(2,2, i + 1)
    randomIndex = random.choice(range(len(trainingData)))
    plt.imshow(trainingData[randomIndex], cmap = plt.cm.binary)
    plt.title(classNames[trainingLabels[randomIndex]])
    plt.axis(False)
```

```python
xTrainScaled.shape
```

```python
# create the model
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax'),
])
#compile the model
model.compile(
    loss= tf.keras.losses.SparseCategoricalCrossentropy,
    optimizer=tf.keras.optimizers.Adam(),
    metrics=['accuracy']
)
# fit the model
model.fit(xTrainScaled, trainingLabels, validation_split= 0.2, epochs=100)
```

```python
# evaluate the image
model.evaluate(xTestScaled, testingLabels)
```

```python
# CNN model
modelFour = tf.keras.Sequential()
modelFour.add (tf.keras.layers.Conv2D(32, kernel_size=(3,3), activation='relu', input_shape=(28,28,1)))
modelFour.add (tf.keras.layers.MaxPooling2D(pool_size=(2,2)))
modelFour.add (tf.keras.layers.Conv2D(64, kernel_size=(3,3), activation='relu')) 
modelFour.add (tf.keras.layers.MaxPooling2D(pool_size=(2,2)))
modelFour.add (tf.keras.layers.Flatten())
modelFour.add (tf.keras.layers.Dense(128, activation ='relu'))
modelFour.add (tf.keras.layers.Dropout (0.5))
modelFour.add (tf.keras.layers.Dense(64, activation='relu')) 
modelFour.add(tf.keras.layers.Dropout(0.5))
modelFour.add (tf.keras.layers.Dense(10, activation='softmax'))

modelFour.compile(loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits = True),
             optimizer=tf.keras.optimizers.Adam(),
             metrics=["accuracy"])

# fit the model for training
modelFour.fit(xTrainScaled, trainingLabels, epochs=50)
```

```python
modelFour.evaluate(xTestScaled, testingLabels)
```
